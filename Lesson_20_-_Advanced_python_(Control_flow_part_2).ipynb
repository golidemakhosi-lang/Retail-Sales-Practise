# Enumerate
add counter to each item

```python
customers = ["Amanda","Jean","James","Ramarumo"]

for index, customer in enumerate(customers, start=1):
    print(index, customer)
```

# Zip
pair items 

```python
stores = ["Hell Kitchen", "Astoria", "Lower Manhattan"]
revenues =[45000,20000,54000]

for store, rev in zip(stores, revenues):
    print(store, rev)
```

```python
import pandas as pd

lst_stores = ["Hell Kitchen", "Astoria", "Lower Manhattan"]
lst_revenues =[45000,20000,54000]
lst_product = ["Coffee","Tea","Mocha"]

df = pd.DataFrame({'Stores': lst_stores, 'Revenue': lst_revenues, 'Product':lst_product})

df.head()
```

```python
for store, rev in zip(df["Stores"], df["Revenue"]):
    print(store, rev)
```

# While loops

- Runs a block of code for as long as a condition is True
- for loop - where you know upfront how many items you're looping over

```python
counter = 1

while counter < 5:
    print(f"counter {counter} Less than 5")
    counter = counter + 1 # without this, the loop runs forever
```

```python
max_attempts = 3
attempts = 0

while attempts < max_attempts:
    print(f"Attempt {attempts+1} of {max_attempts}")
    attempts = attempts +1
```

# Break & Continue

- break - Exits the loop completely
- continue - skips the rest of this iteration

```python
#break
lst_product = ["Coffee","Tea","Mocha", "Chocolate","Bakery"]

for product in lst_product:
    if product == "Chocolate":
        print("Found Chocolate - Stopping search")
        break

    print('Checked: ', product)
```

```python
print(100/0)
```

```python
#continue

Test_sccores = [50, 75, 0,10,80]

for x in Test_sccores:
    if x ==0:
        continue #skip zeros
    print(100/x)
```
