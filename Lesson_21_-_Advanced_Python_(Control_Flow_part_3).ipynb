# List 

- Write a loop in a single line.
- it is faster than a loop

```python
#Loop
sums_ = []

for x in range(5):
    # var = x + x
    sums_.append(x+x)

print(sums_)
```

```python
#list comprehension
sum_1 = [x + x for x in range(5)]
print(sum_1)
```

```python
#loop
grades = [55, 78, 90, 42]
results = []

for g in grades:
    if g >= 60:
        results.append("Pass")
    else:
        results.append("Fail")

print(results)
```

```python
#list comprehension

grades = [55, 78, 90, 42]

results = ["Pass" if g >=60 else "Fail" for g in grades]

print(grades)
print(results)
```

```python
double = lambda x: x*2

print(double(5))
```

```python
def calculate_tier(spend, threshold=500):
    if spend >= threshold *2:
        return "High"
    elif spend >= threshold:
        return "Mid"
    else:
        return "Low"
```

```python
classify = lambda spend: "High" if spend >= 1000 else ("Mid" if spend >= 500 else "Low")

print(classify(750))
```

```python
import pandas as pd

spend = [200, 500, 900, 1200]

df = pd.DataFrame({ "Spend": spend})
df.head()
```

```python
df["Tier"] = df["Spend"].apply(classify)

df.head()
```
