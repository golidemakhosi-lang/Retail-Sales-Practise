# If / elif / else

Making decisions - running different code depending on a condition

monthly_spend=850 --> if >= 1000? -> elif >= 500? -> 'Mid Value' 

```python
monthly_spend = 850

if monthly_spend >= 1000:
    segment = "High Value"
elif monthly_spend >= 500:
    segment = "Mid Value"
else:
    segment = "Low Value"


print(segment)
```

monthly_spend=499 --> if >= 1000? -> elif >= 500? -> else -> "Low Value"

```python
monthly_spend = 499

if monthly_spend >= 1000:
    segment = "High Value"
elif monthly_spend >= 500:
    segment = "Mid Value"
else:
    segment = "Low Value"


print(segment)
```

== (Checks if 2 things are the same)

= (Assigns a value to a variable)

```python
monthly_spend = 900

if monthly_spend >= 300 and monthly_spend <= 500:
    print("Mid value")

# AND returns something if both conditions are true
```

```python
monthly_spend = 900

if monthly_spend >= 300 or monthly_spend <= 500:
    print("Mid value")

# OR returns something if ONE of the conditions is true
```

```python
monthly_spend = 850

if monthly_spend >= 300:
    if monthly_spend <= 500:
        print("Greater than 300 and less than 500")
    else:
        print("Greater than 300 and greater than 500")
else:
    print("Less than 300")
```

## For
For every item in this collection, run this block of code 

```python
customers = ["John","Doe","Sipho","Lerato"]

print(customers[0])
print(customers[1])
print(customers[2])
print(customers[3])
```
