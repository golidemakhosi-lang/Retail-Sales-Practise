```python
# Sales by Category (Matplotlib)
sales_by_category = rs.groupby('Product_Category')['Total_Amount'].sum()
plt.figure()
sales_by_category.plot(kind='bar')
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
# Sales Trend Over Time
#df['Date'] = pd.to_datetime(df['Date'])
sales_trend = rs.groupby('Date')['Total_Amount'].sum()
plt.figure()
sales_trend.plot()
plt.xticks(rotation=45, ha='right')
plt.title('Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python
# Seaborn - Sales by Region
plt.figure()
sns.barplot(data=rs, x='Month', y='Total_Amount')
plt.title('Sales by Month')
plt.show()
```
