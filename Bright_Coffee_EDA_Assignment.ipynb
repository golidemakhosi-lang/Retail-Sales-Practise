## Task 01: Load & Inspect the Dataset
Load the dataset, inspect structure, missing values, and summary statistics.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

df = pd.read_csv('/Volumes/workspace/default/csv/Bright_Coffee_Shop_Sales.csv', sep=';', decimal=',')
print('Shape:', df.shape)
print('\nColumns:')
print(df.columns.tolist())
print('\nData Types:')
print(df.dtypes)
print('\nMissing Values:')
print(df.isnull().sum())
print('\nNumeric Summary:')
print(df.describe())
print('\nCategorical Summary:')
print(df.describe(include='object'))
print('\nProduct Categories:')
print(df['product_category'].value_counts())
print('\nStore Locations:')
print(df['store_location'].value_counts())
```

Observation: The dataset contains over 149,000 transactions with no missing values, making it suitable for analysis.

## Task 02: Feature Engineering & Distributions

```python
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
df['hour'] = pd.to_datetime(df['transaction_time'], format='%H:%M:%S').dt.hour
df['month'] = df['transaction_date'].dt.month
df['revenue'] = df['unit_price'] * df['transaction_qty']
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['unit_price'], kde=True)
plt.title('Distribution of Unit Price')
plt.xlabel('Unit Price')
plt.ylabel('Frequency')
plt.show()
```

```python
store_rev = df.groupby('store_location')['revenue'].sum().sort_values(ascending=False)
store_rev.plot(kind='bar')
plt.title('Total Revenue by Store')
plt.xlabel('Store')
plt.ylabel('Revenue')
plt.show()
```

```python
hourly = df.groupby('hour')['transaction_id'].count()
hourly.plot(kind='bar')
plt.title('Transactions by Hour')
plt.xlabel('Hour')
plt.ylabel('Transactions')
plt.show()
```

## Task 03: Correlation & Group Analysis

```python
corr = df[['unit_price','transaction_qty','revenue']].corr()
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='Blues')
plt.title('Correlation Heatmap')
plt.show()
```

```python
pivot = pd.pivot_table(df, values='revenue', index='store_location', columns='product_category', aggfunc='sum')
print(pivot)
pivot.plot(kind='bar', figsize=(12,6))
plt.title('Revenue by Store and Product Category')
plt.ylabel('Revenue')
plt.show()
```

## Task 04: Visualisation Library Comparison

```python
cat_rev = df.groupby('product_category')['revenue'].sum().reset_index()

# Matplotlib
plt.figure(figsize=(8,5))
plt.bar(cat_rev['product_category'], cat_rev['revenue'])
plt.title('Revenue by Product Category')
plt.xticks(rotation=45)
plt.show()

# Seaborn
plt.figure(figsize=(8,5))
sns.barplot(data=cat_rev, x='product_category', y='revenue')
plt.title('Revenue by Product Category')
plt.xticks(rotation=45)
plt.show()

# Plotly
fig = px.bar(cat_rev, x='product_category', y='revenue', title='Revenue by Product Category')
fig.show()
```

Seaborn required less code than Matplotlib and produced a more polished chart by default. Matplotlib offers greater customization but is more verbose. Plotly provided the most interactive output through hover labels and zooming. For dashboards and business presentations, Plotly is often the preferred choice because users can explore the data interactively.
