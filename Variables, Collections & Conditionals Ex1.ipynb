Task 3.3 — Write an if/elif/else statement

```python
#Task 3.3  Write your if/elif/else code below and record the output for each test:
#ae.Output when score = 72: ___________
#af.Output when score = 45: ___________
#Output when score = 91
#  Test Code
for test_score in [72, 45, 91]:
    if test_score >= 75:
        grade = "Distinction"
    elif test_score >= 50:
        grade = "Pass"
    else:
        grade = "Fail"
    print(f"Score {test_score}: {grade}")
```
