# Exploratory Data Analysis

Before doing anything with data, you need to understand what's in it. Pandas gives you quick summary tools.

# Loading data

It's like opening a spreadsheet, but with a single line of code instead of double-clicking a file.

```python
df = pd.read_csv("retail_sales_dataset.csv")

df.head()
```

```python
#how many rows and columns
df.shape
```

# Explore Data

```python
#column names, data types, nulls
df.info()
```

```python
# stats: mean, min, max etc
df.describe()
```

```python
df.columns
```

```python
print(df["Total Amount"].sum())
```

```python
print(df["Total Amount"].mean())
```

Selecting Columns and rows

```python
df["Product Category"]
```

```python
# Selecting mutiple columns
# df[["Product Category","Price per Unit"]]

access_column = ["Product Category","Price per Unit"]

print(df[access_column])
```

```python
#selecting rows by position - like a list index
df.iloc[0] #first row
```

```python
df.iloc[0:3] #first 3 rows
```

```python
 df["Age"] < 20 
```

```python
#select row by condition

# df[df["Age"] < 20]

conditions = df["Age"] < 20 

df[conditions]
```

# simple summaries & grouping

```python
#count how many transactions are in each product category
df["Product Category"].value_counts()
```

```python
#average total amount by Product Category

df.groupby("Product Category")["Total Amount"].mean()
```
